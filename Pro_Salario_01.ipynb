```python
""" Aquí van las horas trabajadas en el mes"""

HorasTrabajadas = int(input("Escribe las horas que trabajaste en el mes"))
print (HorasTrabajadas)
```

```python
""" Aquí va el salario por hora"""

SalarioHora = int(input("Escribe tu salario por hora trabajada"))
print(SalarioHora)
```

```python
#SE EJECUTA LA OPERACIÓN QUE CALCULA TU SALARIO
Salario = HorasTrabajadas * SalarioHora
print(f"Tu salario es de: " + str(Salario))
```

```python
#¿TU RENDIMIENTO EN EL MES FUE BUENO?
#AQUÍ TE VA UN AUMENTO SI FUE ASÍ

if HorasTrabajadas >= 100:
  AumentoSalarial = Salario * 0.10
  print("Tu rendimiento es de un 10%")
  #print(AumentoSalarial)
elif HorasTrabajadas < 100:
  AumentoSalarial = Salario * 0.05
  print("Tu rendimiento es de un 5%")
  #print(AumentoSalarial)
else:
  print("No tienes ningun aumento")

```

```python
Nombre = input("Escribe tu nombre")

"""Tu salario final es..."""

SalarioFinal = Salario + AumentoSalarial
print("El salario final de " + Nombre + " es de " + str(SalarioFinal))
```
